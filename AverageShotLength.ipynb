```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.style as style
import seaborn as sns
import scipy.stats as st


data_url = 'merged_purged.csv'

df = pd.read_csv(data_url)
```

```python
##np.corrcoef(df['Year'], df['ASL'])
#Line Chart - Average Shot Length
sns.set_context('paper')
style.use('ggplot')
sns.lineplot(x="Year", y="ASL", data = df)
plt.savefig('ASLbyYear.png', transparent=False)
```

```python
sns.set_context('paper')
style.use('ggplot')
sns.lineplot(x="Year", y="duration", data = df)
plt.savefig('DurationbyYear.png', transparent=False)
```

```python
sns.set_context('paper')
style.use('ggplot')
sns.jointplot(x="Year", y="ASL", data = df)
plt.savefig('DataOverview.png', transparent=False)
```
